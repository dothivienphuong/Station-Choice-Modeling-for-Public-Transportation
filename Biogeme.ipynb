```python
import pandas as pd
import biogeme.database as db
import biogeme.biogeme as bio
from biogeme.expressions import Beta, Variable
import biogeme.models as models
import warnings

# 1. Load data
df = pd.read_excel("dataset_output.xlsx")
df = df.dropna(subset=['id']).copy()

# Map choices 
choice_mapping = {
    "opt_out": 0, 
    "A": 1, "B": 2, "C": 3, "D": 4, "E": 5, "F": 6, 
    "G": 7, "H": 8, "I": 9, "J": 10, "K": 11, "L": 12
}
df["choice_numeric"] = df["choice"].map(choice_mapping)
df_numeric = df.drop(columns=["choice"])

database = db.Database("station_choice_dataset", df_numeric)

# 2. Define Alternative-Specific Constants (ASCs)
ASC_OPT_OUT = Beta('ASC_OPT_OUT', 0, None, None, 0)
ASC_A = Beta('ASC_A', 0, None, None, 1)  # Fixed reference station
ASC_B = Beta('ASC_B', 0, None, None, 0)
ASC_C = Beta('ASC_C', 0, None, None, 0)
ASC_D = Beta('ASC_D', 0, None, None, 0)
ASC_E = Beta('ASC_E', 0, None, None, 0)
ASC_F = Beta('ASC_F', 0, None, None, 0)
ASC_G = Beta('ASC_G', 0, None, None, 0)
ASC_H = Beta('ASC_H', 0, None, None, 0)
ASC_I = Beta('ASC_I', 0, None, None, 0)
ASC_J = Beta('ASC_J', 0, None, None, 0)
ASC_K = Beta('ASC_K', 0, None, None, 0)
ASC_L = Beta('ASC_L', 0, None, None, 0)

# Attributes and Interactions
B_DIST = Beta("B_DIST", 0, None, None, 0)
B_SHELTER = Beta("B_SHELTER", 0, None, None, 0)
B_RAINY = Beta("B_RAINY", 0, None, None, 0)           
B_DIST_RAINY = Beta("B_DIST_RAINY", 0, None, None, 0) 

choice_numeric = Variable("choice_numeric")
rainy = Variable("rainy")

# 3. Utilities
V0 = ASC_OPT_OUT

V1 = ASC_A + B_DIST * Variable("distance_A") + B_SHELTER * Variable("shelter_A") + B_RAINY * rainy + B_DIST_RAINY * Variable("distance_A") * rainy
V2 = ASC_B + B_DIST * Variable("distance_B") + B_SHELTER * Variable("shelter_B") + B_RAINY * rainy + B_DIST_RAINY * Variable("distance_B") * rainy
V3 = ASC_C + B_DIST * Variable("distance_C") + B_SHELTER * Variable("shelter_C") + B_RAINY * rainy + B_DIST_RAINY * Variable("distance_C") * rainy
V4 = ASC_D + B_DIST * Variable("distance_D") + B_SHELTER * Variable("shelter_D") + B_RAINY * rainy + B_DIST_RAINY * Variable("distance_D") * rainy
V5 = ASC_E + B_DIST * Variable("distance_E") + B_SHELTER * Variable("shelter_E") + B_RAINY * rainy + B_DIST_RAINY * Variable("distance_E") * rainy
V6 = ASC_F + B_DIST * Variable("distance_F") + B_SHELTER * Variable("shelter_F") + B_RAINY * rainy + B_DIST_RAINY * Variable("distance_F") * rainy
V7 = ASC_G + B_DIST * Variable("distance_G") + B_SHELTER * Variable("shelter_G") + B_RAINY * rainy + B_DIST_RAINY * Variable("distance_G") * rainy
V8 = ASC_H + B_DIST * Variable("distance_H") + B_SHELTER * Variable("shelter_H") + B_RAINY * rainy + B_DIST_RAINY * Variable("distance_H") * rainy
V9 = ASC_I + B_DIST * Variable("distance_I") + B_SHELTER * Variable("shelter_I") + B_RAINY * rainy + B_DIST_RAINY * Variable("distance_I") * rainy
V10 = ASC_J + B_DIST * Variable("distance_J") + B_SHELTER * Variable("shelter_J") + B_RAINY * rainy + B_DIST_RAINY * Variable("distance_J") * rainy
V11 = ASC_K + B_DIST * Variable("distance_K") + B_SHELTER * Variable("shelter_K") + B_RAINY * rainy + B_DIST_RAINY * Variable("distance_K") * rainy
V12 = ASC_L + B_DIST * Variable("distance_L") + B_SHELTER * Variable("shelter_L") + B_RAINY * rainy + B_DIST_RAINY * Variable("distance_L") * rainy

# Utility and Availability 
V = {0: V0, 1: V1, 2: V2, 3: V3, 4: V4, 5: V5, 6: V6, 7: V7, 8: V8, 9: V9, 10: V10, 11: V11, 12: V12}
av = {0: 1, 1: 1, 2: 1, 3: 1, 4: 1, 5: 1, 6: 1, 7: 1, 8: 1, 9: 1, 10: 1, 11: 1, 12: 1}

# 4. Estimate
logprob = models.loglogit(V, av, choice_numeric)
biogeme_model = bio.BIOGEME(database, logprob)
biogeme_model.model_name = "final_mip_model"
results = biogeme_model.estimate()

with warnings.catch_warnings():
    warnings.simplefilter("ignore", DeprecationWarning)
    print(results.get_estimated_parameters().iloc[:, 0:2])
```
