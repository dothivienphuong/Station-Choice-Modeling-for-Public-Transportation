```python
import matplotlib.pyplot as plt

budget_limit = [4, 5, 6, 7, 8, 9, 10, 11, 12]
total_transit_share = [61.67, 66.05, 69.48, 72.19, 74.34, 76.18, 77.72, 78.95, 80.00]  

# Initialize the plot
plt.figure(figsize=(8, 5))

# Create the line plot 
plt.plot(budget_limit, total_transit_share, marker='o', linestyle='-', color='b', linewidth=2, label='Transit share')

# Set axes limits and labels
plt.xlim(4, 12)
plt.xlabel('Budget limit')
plt.ylabel('Total transit share (%)')

# Add styling elements
plt.title('Total transit share across budget limits')
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()

# Display the graph
plt.show()
```
