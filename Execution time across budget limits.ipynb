```python
import matplotlib.pyplot as plt

budget_limit = [4, 5, 6, 7, 8, 9, 10, 11, 12]
execution_time = [173.9, 283.9, 248.0, 90.0, 56.3, 19.4, 6.8, 3.2, 2.6]  

# Initialize the plot
plt.figure(figsize=(8, 5))

# Create the line plot 
plt.plot(budget_limit, execution_time, marker='o', linestyle='-', color='b', linewidth=2, label='Execution time')

# Set axes limits and labels
plt.xlim(4, 12)
plt.xlabel('Budget Limit')
plt.ylabel('Execution time (seconds)')

# Add styling elements
plt.title('Execution time across budget limits')
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()

# Display the graph
plt.show()
```
